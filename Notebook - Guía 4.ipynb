### Análisis de sensibilidad entre tasa de descuento y cantidades demandadas de un bien

```python
import numpy as np
import matplotlib.pyplot as plt

x = df['cantidad']
y = df['porcentaje descuento']

# Calculamos los coeficientes de la regresión lineal
coeficientes = np.polyfit(x, y, 1)
polinomio = np.poly1d(coeficientes)

# Creamos el gráfico de dispersión
plt.scatter(x, y)

# Graficamos la línea de tendencia
xx = np.linspace(min(x), max(x), 100)
yy = polinomio(xx)
plt.plot(xx, yy, color='red')

plt.title('Gráfico de dispersión entre cantidad y porcentaje descuento')
plt.xlabel('cantidad')
plt.ylabel('porcentaje descuento')

plt.show()
```

```python
import statsmodels.api as sm

X = df['porcentaje descuento']
y = df['cantidad']

# Añadimos una constante a la variable explicativa
X = sm.add_constant(X)

# Creamos el modelo de MCO
modelo = sm.OLS(y, X)

# Ajustamos el modelo
resultados = modelo.fit()

# Imprimimos los resultados
print(resultados.summary())
```

Evidenciamos que a partir de un modelo de Mínimos Cuadrados Ordinarios (MCO), el coeficiente para ‘porcentaje descuento’ es 5.4632. Esto significa que, en promedio, por cada aumento de una unidad en ‘porcentaje descuento’, la ‘cantidad’ aumenta en 5.4632 unidades, manteniendo todo lo demás constante. Así mismo, para determinar si las variables son significativas para el modelo, puedes mirar el valor p asociado a cada variable (en la columna ‘P>|t|’). Un valor p pequeño (generalmente, menor a 0.05) indica que puedes rechazar la hipótesis nula de que el coeficiente de esa variable es cero en la población. En otras palabras, un valor p pequeño sugiere que la variable es significativa para el modelo.

### Análisis de elasticidad cruzadas de los productos manejados

```python
import pandas as pd

df['fecha'] = pd.to_datetime(df['fecha'])
df_agregado = df.groupby(['fecha', 'producto']).agg({'cantidad': 'sum', 'precio': 'mean'}).reset_index()

# Creamos un DataFrame para las cantidades
df_cantidad = df_agregado.pivot(index='fecha', columns='producto', values='cantidad')

# Creamos un DataFrame para los precios
df_precio = df_agregado.pivot(index='fecha', columns='producto', values='precio')

# Ahora 'df_cantidad' y 'df_precio' son DataFrames que contienen las cantidades demandadas y los precios de todos los productos a lo largo del tiempo
```

```python
import statsmodels.formula.api as smf
import itertools

# Obtenemos la lista de productos
productos = df_cantidad.columns

# Creamos un diccionario para almacenar los resultados
elasticidades = {}

# Iteramos sobre todas las combinaciones de productos
for producto1, producto2 in itertools.combinations(productos, 2):
    # Creamos un DataFrame temporal con las cantidades y precios de los dos productos
    df_temp = pd.DataFrame({
        'cantidad_producto1': df_cantidad[producto1],
        'precio_producto1': df_precio[producto1],
        'precio_producto2': df_precio[producto2]
    }).dropna()

    # Verificamos si df_temp está vacío
    if not df_temp.empty:
        # Ajustamos el modelo de regresión
        modelo = smf.ols(formula='cantidad_producto1 ~ precio_producto1 + precio_producto2', data=df_temp)
        resultados = modelo.fit()

        # Guardamos la elasticidad cruzada en el diccionario
        elasticidades[(producto1, producto2)] = resultados.params['precio_producto2']

# Imprimimos las elasticidades cruzadas
for productos, elasticidad in elasticidades.items():
    print('La elasticidad cruzada de la demanda entre {} y {} es: {}'.format(productos[0], productos[1], elasticidad))
```

### Interpretación
La elasticidad cruzada de la demanda ayuda a proporcionar una intuición sobre si un par de productos son sustitutos, complementarios o son indiferentes entre sí sobre los patrones de consumo. Así las cosas, la elasticidad puede tomar valores:

- Elasticidad cruzada > 0: Los productos son sustitutos. Un aumento en el precio de un producto lleva a un aumento en la demanda del otro producto.

- Elasticidad cruzada < 0: Los productos son complementarios. Un aumento en el precio de un producto lleva a una disminución en la demanda del otro producto.

- Elasticidad cruzada = 0: Los productos son independientes. Los cambios en el precio de un producto no afectan a la demanda del otro producto.


Así, por ejemplo, la elasticidad cruzada de la demanda entre Aceite Naoli x 20Lt Bidón por 20 litros Unidad y Ajo Estandar Kg es: 0.04371566519090225. Lo que significa que dicho par de productos son sustitutos.
